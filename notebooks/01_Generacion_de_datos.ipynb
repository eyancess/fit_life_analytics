```python
def visitas_base_mes():

    return np.random.choice(
        [4, 6, 8, 10, 12, 14, 16, 18],
        p=[0.05, 0.10, 0.15, 0.25,
           0.20, 0.15, 0.07, 0.03]
    )
```

```python
for _ in range(10):
    print(visitas_base_mes())
```

```python
#visitas_mes_socio
def calcular_visitas_mes(
    visitas_base,
    churn_final,
    meses_hasta_baja
):

    visitas = visitas_base

    if churn_final == 1:

        if meses_hasta_baja <= 1:
            visitas *= 0.20

        elif meses_hasta_baja <= 2:
            visitas *= 0.40

        elif meses_hasta_baja <= 3:
            visitas *= 0.60

        elif meses_hasta_baja <= 4:
            visitas *= 0.80

    return max(0, round(visitas))
```

```python
for meses in [6, 4, 3, 2, 1]:
    print(
        meses,
        calcular_visitas_mes(
            visitas_base=12,
            churn_final=1,
            meses_hasta_baja=meses
        )
    )
```

```python
## Tabla Temporal mensual para cada socio
registros_socio_mes = []

for _, socio in dim_socios.iterrows():

    fecha_inicio = socio["fecha_alta"]
    fecha_fin = socio["fecha_baja"] if pd.notna(socio["fecha_baja"]) else pd.Timestamp(FECHA_FIN)

    meses = pd.period_range(
        start=fecha_inicio.to_period("M"),
        end=fecha_fin.to_period("M"),
        freq="M"
    )

    visitas_base = visitas_base_mes()

    for mes_periodo in meses:

        mes = mes_periodo.to_timestamp()

        meses_hasta_baja = 999

        if pd.notna(socio["fecha_baja"]):
            meses_hasta_baja = (
                (socio["fecha_baja"].year - mes.year) * 12
                + (socio["fecha_baja"].month - mes.month)
            )

        visitas_mes = calcular_visitas_mes(
            visitas_base=visitas_base,
            churn_final=socio["churn_final"],
            meses_hasta_baja=meses_hasta_baja
        )

        registros_socio_mes.append({
            "id_socio": socio["id_socio"],
            "id_centro": socio["id_centro"],
            "mes": mes,
            "visitas_mes": visitas_mes
        })

fact_socio_mes_temp = pd.DataFrame(registros_socio_mes)

fact_socio_mes_temp.head()
```

```python
def generar_hora_entrada():
    return np.random.choice(
        ["07:00", "08:00", "09:00", "13:00", "14:00", "18:00", "19:00", "20:00"],
        p=[0.08, 0.10, 0.07, 0.08, 0.07, 0.25, 0.22, 0.13]
    )


def generar_duracion():
    duracion = np.random.normal(loc=65, scale=20)
    return int(np.clip(round(duracion), 20, 180))


def generar_tipo_actividad():
    return np.random.choice(
        ["sala_libre", "cardio", "clase_grupal"],
        p=[0.60, 0.25, 0.15]
    )
```

```python
visitas = []
id_visita = 1

for _, fila in fact_socio_mes_temp.iterrows():

    n_visitas = int(fila["visitas_mes"])
    
    for _ in range(n_visitas):
        
        dia = np.random.randint(1, 29)
        fecha_visita = pd.Timestamp(
            year=fila["mes"].year,
            month=fila["mes"].month,
            day=dia
        )
        
        id_fecha = int(fecha_visita.strftime("%Y%m%d"))
        
        visitas.append({
            "id_visita": id_visita,
            "id_socio": fila["id_socio"],
            "id_centro": fila["id_centro"],
            "id_fecha": id_fecha,
            "fecha_visita": fecha_visita,
            "hora_entrada": generar_hora_entrada(),
            "duracion_minutos": generar_duracion(),
            "tipo_actividad": generar_tipo_actividad()
        })
        
        id_visita += 1

fact_visitas = pd.DataFrame(visitas)

fact_visitas.head()
```
